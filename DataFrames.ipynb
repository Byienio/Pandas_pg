# Add and remove columns

```python
fuel = pd.read_csv('data/data/fuel.csv',low_memory=False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'],index_col='Vehicle ID')
fuel.head()
```

```python
fuel['Combined KPL'] = fuel['Combined MPG (FT1)'] * 1.609 / 3.78
fuel.head()
```

```python
fuel.insert(loc=5,column = 'liters per km',value=100/fuel['Combined KPL'])
fuel.head()
```

```python
del fuel['Combined MPG (FT1)']
fuel.head()
```

```python
fuel.drop('Combined KPL',axis=1,inplace=True)
fuel.head()
```

# Clearing data with fillna

```python
fuel = pd.read_csv('data/data/fuel.csv',low_memory=False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'],index_col='Vehicle ID')
fuel.loc[27705, 'Class'] = np.nan
fuel.loc[26561, 'Class'] = np.nan
fuel.loc[27550, 'Fuel Type'] = np.nan
fuel.loc[27705, 'Combined MPG (FT1)'] = np.nan
fuel.loc[27681, 'Combined MPG (FT1)'] = np.nan
fuel.head()
```

```python
fuel.fillna(-1).head()
```

```python
replaceRules = {'Class':'---','Fuel Type':'---','Combined MPG (FT1)':-1}
```

```python
fuel.fillna(replaceRules).head()
```

```python
avgMPG = fuel['Combined MPG (FT1)'].mean()
```

```python
fuel.fillna({'Class':'?', 'Fuel Type':'?', 'Combined MPG (FT1)':avgMPG},inplace=True)
fuel.head()
```

```python
fuel['Combined MPG (FT1)']= fuel['Combined MPG (FT1)'].ffill()
fuel.head()
```

# Ranking

```python
fortune = pd.read_csv('data/data/Fortune_500_2017.csv',usecols=['Rank','Title','Employees','Profits','Assets'],index_col='Rank')
fortune.head()
```

```python
fortune['Employees'].rank( ascending=False).head()
```

```python
fortune['RankByEmployee'] = fortune['Employees'].rank( ascending=False)
fortune.head()
```

```python
fortune['Profits'].rank( ascending=False).head()
```

```python
fortune['RankByProfit'] = fortune['Profits'].rank( ascending=False)
fortune.head()
```

```python
fortune.nlargest(3, 'Assets')
```

```python
fortune.nsmallest(3, 'Assets')
```

# Filtering DataFrames

```python
isEmployeesRankFirst10 = fortune['RankByEmployee'] <= 10
```

```python
isProfitRankFirst10 = fortune['RankByProfit'] <= 10
```

```python
fortune[isEmployeesRankFirst10 & isProfitRankFirst10]
```

```python
isEmployeesRankMore400 = fortune['RankByEmployee'] >= 400
```
