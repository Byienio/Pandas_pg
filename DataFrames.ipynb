# Clearing data with fillna

```python
fuel = pd.read_csv('data/data/fuel.csv',low_memory=False, usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)'],index_col='Vehicle ID')
fuel.loc[27705, 'Class'] = np.nan
fuel.loc[26561, 'Class'] = np.nan
fuel.loc[27550, 'Fuel Type'] = np.nan
fuel.loc[27705, 'Combined MPG (FT1)'] = np.nan
fuel.loc[27681, 'Combined MPG (FT1)'] = np.nan
fuel.head()
```

```python
fuel.fillna(-1).head()
```

```python
replaceRules = {'Class':'---','Fuel Type':'---','Combined MPG (FT1)':-1}
```

```python
fuel.fillna(replaceRules).head()
```

```python
avgMPG = fuel['Combined MPG (FT1)'].mean()
```

```python
fuel.fillna({'Class':'?', 'Fuel Type':'?', 'Combined MPG (FT1)':avgMPG},inplace=True)
fuel.head()
```

```python
fuel['Combined MPG (FT1)']= fuel['Combined MPG (FT1)'].ffill()
fuel.head()
```

# isin, isnull, notnull, between

```python
fuel = pd.read_csv("data/data/fuel.csv", usecols=['Vehicle ID','Year','Make', 'Model','Class','Fuel Type',
'Combined MPG (FT1)','Start Stop Technology'], index_col = 'Vehicle ID', low_memory=False)
fuel.head()
```

```python
listOfCars = ['Renault','Toyota','Ford']
isInList = fuel['Make'].isin(listOfCars)
isInList.head()
```

```python
fuel[isInList].head()
```

```python
isNoStartStopDefined = fuel['Start Stop Technology'].isnull()
isNoStartStopDefined.head()
```

```python
fuel[isNoStartStopDefined].head()
```

```python
isStartStopDefined = fuel['Start Stop Technology'].notnull()
isStartStopDefined.head()
```

```python
fuel[isStartStopDefined].head()
```

```python
mpg50_60 = fuel['Combined MPG (FT1)'].between(50,60)
mpg50_60.head()
```

```python
fuel[mpg50_60].head()
```

```python
print(f"There are {len(fuel[isInList])} cars within the list ['Renault', 'Toyota', 'Ford']")
print(f"There are {len(fuel[isNoStartStopDefined])} cars without Start Stop Technology defined")
print(f"There are {len(fuel[mpg50_60])} cars with Combined MPG between 50 and 60")
print(f"There are {len(fuel[isStartStopDefined])} cars with Start Stop Technology defined")
```
