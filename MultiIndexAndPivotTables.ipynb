# Swaplevel

```python
fuel = pd.read_csv('data/data/fuel.csv',usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'],index_col=['Year','Make','Model'])
fuel.sort_index(inplace=True)
fuel.head()
```

```python
fuel.loc[(1984,'Alfa Romeo','Spider Veloce 2000')]
```

```python
fuel = fuel.swaplevel(0,1).sort_index()
fuel.head()
```

```python
fuel.loc[('Alfa Romeo',1984,'Spider Veloce 2000')]
```

```python
fuel= fuel.swaplevel(1,2).sort_index()
fuel.head()
```

```python
fuel.loc[('Alfa Romeo','Spider Veloce 2000',1984)]
```

# Pivot tables

```python
from datetime import datetime,timedelta
import time
```

```python
df = pd.read_csv('data/data/marathon_results_2016.csv',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'],index_col='Bib')
```

```python
df['40K'] = df['40K'].replace('-', 0)
df['Half'] = df['Half'].replace('-', 0)

df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)

df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))

df['TotalSeconds'] = df['TotalSeconds'].replace(0,np.nan)
df['HalfSeconds'] = df['HalfSeconds'].replace(0,np.nan)
```

```python
df.head()
```

```python
df.pivot_table(index='Age',columns='M/F',values='TotalSeconds',aggfunc='mean')
```

```python
pivot_df= df.pivot_table(index='Age',columns='M/F',values=['TotalSeconds','HalfSeconds'],aggfunc='mean')
pivot_df.head()
```

```python
pivot_df.swaplevel(axis='columns').head()
```

```python
pivot_df.swaplevel(axis='columns').sort_index(axis=1).head()
```

```python
df.pivot_table(index='Age',columns='M/F',values='Country',aggfunc='count').head()
```

```python
df.pivot_table(index='Age',columns='M/F',values='Country',aggfunc='count').head()
```

```python
df.pivot_table(index='Country',values='City',aggfunc='count')
```

```python
df.pivot_table(index='Country',values='City',aggfunc='count').sort_values(by='City',ascending=False)
```

```python
df.pivot_table(index=['Country','State','City'],values='TotalSeconds',aggfunc='mean')
```

```python
df.pivot_table(index=['Country','State','City'],values='TotalSeconds',aggfunc='mean').sort_values(by='TotalSeconds',ascending=False)
```

# Melt

```python
df = pd.read_csv('data/data/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta, errors='coerce')
df['Half'] = df['Half'].apply(pd.to_timedelta, errors='coerce')
df = df[df['40K'].notna() & df['Half'].notna()]
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values="TotalSeconds").head()
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.melt(id_vars='Age')
```

```python
df.melt(id_vars='Age',value_name='TotalSeconds')
```

```python
df.melt(id_vars='Age',value_name='TotalSeconds',var_name='Sex')
```

```python
df = pd.read_csv('data/data/marathon_results_2016.csv', index_col='Bib',
usecols=
['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta, errors='coerce')
df['Half'] = df['Half'].apply(pd.to_timedelta, errors='coerce')
df = df[df['40K'].notna() & df['Half'].notna()]
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values=
["HalfSeconds","TotalSeconds"]).head()
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.columns = df.columns.to_flat_index().str.join('')
df.head()
```

```python
df.melt(id_vars='Age')
```

```python
df.melt(id_vars='Age', value_name='Time')
```
