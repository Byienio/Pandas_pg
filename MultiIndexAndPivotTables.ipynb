# Transposing

```python
fuel = pd.read_csv('data/data/fuel.csv',usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'],index_col=['Year','Make','Model']).sort_index()
fuel.head()
```

```python
fuel.info()
```

```python
fuel.loc[(1984)]
```

```python
fuel.loc[(1984,'Alfa Romeo')]
```

```python
fuel.loc[(1984,'Alfa Romeo','Spider Veloce 2000')]
```

```python
fuel.loc[ (1984,'Alfa Romeo','Spider Veloce 2000'),'Range (FT1)' ]
```

```python
fuel = fuel.transpose()
```

```python
fuel.head()
```

```python
fuel.loc['Range (FT1)'].head()
```

```python
fuel.loc['Range (FT1)',(1984)]
```

```python
fuel.loc['Range (FT1)',(1984,'Alfa Romeo')]
```

```python
fuel.loc['Range (FT1)',(1984,'Alfa Romeo','Spider Veloce 2000')]
```

# Swaplevel

```python
fuel = pd.read_csv('data/data/fuel.csv',usecols=['Year','Make','Model','Range (FT1)','City Range (FT1)'],index_col=['Year','Make','Model'])
fuel.sort_index(inplace=True)
fuel.head()
```

```python
fuel.loc[(1984,'Alfa Romeo','Spider Veloce 2000')]
```

```python
fuel = fuel.swaplevel(0,1).sort_index()
fuel.head()
```

```python
fuel.loc[('Alfa Romeo',1984,'Spider Veloce 2000')]
```

```python
fuel= fuel.swaplevel(1,2).sort_index()
fuel.head()
```

```python
fuel.loc[('Alfa Romeo','Spider Veloce 2000',1984)]
```
