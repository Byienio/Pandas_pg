```python
mcdonalds = pd.read_csv('data/data/Mcdonalds.csv',usecols=['Category','Item','Calories','TotalFat'])
mcdonalds.head()
```

# Working with Excel

```python
mcdonalds = pd.read_csv('data/data/Mcdonalds.csv',usecols=['Category','Item','Calories','TotalFat'])
mcdonalds.head()
```

```python
with pd.ExcelWriter("exported_data/mcdonalds-excel.xlsx", mode="w") as excelWriter:
    mcdonalds.to_excel(excelWriter, sheet_name="Menu",index=False)
```

```python
imported1 = pd.read_excel('exported_data/mcdonalds-excel.xlsx',sheet_name='Menu')
imported1.head()
```

```python
imported1['Calories'] = imported1['Calories']/2
```

```python
with pd.ExcelWriter('exported_data/mcdonalds-excel.xlsx', mode='a') as excelWriter:
    imported1.to_excel(excelWriter,sheet_name='Menu Light')
```

```python
imported1['Calories'] = imported1['Calories']/2
```

```python
with pd.ExcelWriter('exported_data/mcdonalds-excel.xlsx', mode='a') as excelWriter:
    imported1.to_excel(excelWriter,sheet_name='Menu Extra Light')
```
