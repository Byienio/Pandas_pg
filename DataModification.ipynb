# Editing DataFrame index and columns

```python
import pandas as pd
import numpy as np
```

```python
professions = pd.read_csv('data/data/Prestige.csv',index_col='name')
professions.head()
```

```python
professionsOriginal = professions.copy()
```

```python
professions.rename(columns={'income': 'incomeDollarsPerYear'},inplace=True)
professions.head()
```

```python
professions.rename(columns={'women':'womenPercent','prestige':'prestigeIndicator'},inplace=True)
professions.head()
```

```python
professions.rename(index={'accountants':'bookkeepers','general.managers':'managers'},inplace=True)
professions.head()
```

# Data modification

```python
professions = pd.read_csv('data/data/Prestige.csv', index_col='name')
professions.head()
```

# Adding and removing rows

```python
professions = pd.read_csv('data/data/Prestige.csv')
professions.head()
```

```python
dictionary = {'name':'[data scientist]','education':[np.nan] ,'income':[np.nan],'women':[np.nan],'prestige':[np.nan],'type':[np.nan]}
dictionary
```

```python
professions = pd.concat([professions,pd.DataFrame(dictionary)])
professions.tail()
```

```python
taxi = professions[professions['name'] == 'taxi.drivers']
taxi
```

```python
professions.drop(axis='rows',labels=98,inplace=True)
professions.tail()
```

```python
professions = pd.concat([professions, taxi])
professions.tail()
```

```python
professions.drop(axis='columns',columns = 'census',inplace=True)
professions.head()
```

```python
del professions['type']
professions.head()
```
